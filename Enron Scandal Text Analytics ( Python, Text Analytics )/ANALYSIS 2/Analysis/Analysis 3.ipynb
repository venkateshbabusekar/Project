- For the given year 2016 we are analysing the views and shares for past 30 days

```python
# Importing the important package  
import glob
import os
import json
from operator import itemgetter
from os.path import isfile,isdir
import datetime
import matplotlib.pyplot as plt
import csv
import matplotlib.pyplot as plt
```

```python
# we are finding the path for the file 
sfile = glob.glob(os.path.join(os.path.join(os.getcwd(), 'Data_processed\\Analysis - 3 _Data\\Shared'), '*.json'))
vfile = glob.glob(os.path.join(os.path.join(os.getcwd(), 'Data_processed\\Analysis - 3 _Data\\Viewed'), '*.json'))
```

```python
#We are defining the length of month 
monthDays = list(range(1,31))
```

```python
#We are creating the dictionaty and list 
shareddic = {}
xs = []
ys = []
viewsDict = {}
xv=[]
yv=[]
```

```python
# We are creating the loop for the sharedfile  
for f in sfile:
        with open(f,'r') as json_file:
            json_file = json.load(json_file) # loading the json file 
            final = json_file['results'] # finding the the result 
            for share in final: # looping all the 
                date = share['published_date']  # finding the publised data
                total_shares = share['total_shares'] # finding the total shared
                Day = datetime.datetime.strptime(date,'%Y-%m-%d')
                day = Day.day
                months = Day.month
                if months == 2:
                    if day not in shareddic: # we are adding the value to dictionary 
                        shareddic[day] = total_shares
                    else:
                        shareddic[day] += total_shares
                    for each in monthDays: 
                        if each != day:
                            if each not in shareddic:  # we are adding the value to dictionary 
                                shareddic[each] = 0
```

```python
# adding the value to the list 
xs = list(shareddic.keys())
ys = list(shareddic.values())
```

```python
#Plotting the graph 
fig, ax = plt.subplots(figsize=(10,5))
plt.bar(xs,ys,width = 0.50)
ax.set_ylabel('Number of Shares')
ax.set_xlabel('For the month of Feb 2017') 
plt.xlim(1,30)
plt.ylim(10,90)
ax.set_title('Share details for the year 2017')
width = 1
ax.set_xticks([p+width for p in xs])

plt.grid()
plt.show()
```

```python
# We are creating the loop for the sharedfile  
for file in vfile:
        with open(file,'r') as json_file:
            json_file = json.load(json_file)# loading the json file 
            results = json_file['results'] # finding the the result 
            for views in results:
                Date = views['published_date'] # finding the the published_date 
                total_views = views['views']
                Day = datetime.datetime.strptime(Date,'%Y-%m-%d')
                day = Day.day
                month = Day.month
                if month == 2:
                    if day not in viewsDict: # we are adding the value to dictionary 
                        viewsDict[day] = total_views
                    else:
                        viewsDict[day] += total_views
                    for each in monthDays:
                        if each != day:
                            if each not in viewsDict: # we are adding the value to dictionary 
                                viewsDict[each] = 0
            
```

```python
# adding the value to the list 
xv = list(viewsDict.keys())
yv = list(viewsDict.values())
```

```python
#Plotting the graph 
fig, ax = plt.subplots(figsize=(10,5))
plt.bar(xv,yv,width = 0.50)
ax.set_ylabel('Number of Views')# Set the x axis label
ax.set_xlabel('For the month of Feb 2017') #Setting the x-axis and y-axis limits
plt.xlim(1,30)
plt.ylim(10,90)# Set the chart's title
ax.set_title('View details for the year 2017')
width = 1
ax.set_xticks([p+width for p in xv])

plt.grid()
plt.show()
```
