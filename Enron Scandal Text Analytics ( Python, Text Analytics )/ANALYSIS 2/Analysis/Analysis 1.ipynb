# Finding 1 

- For the files dated from 1940 to 1945 we are finding how many words that a related to war and over the period of time how the frequency and the words are distributed

```python
#We importing the necessary package 
import json
import glob
import os
import re
from nltk.tokenize import sent_tokenize, word_tokenize
import matplotlib.pyplot as plt
```

```python
# file path 
filepath =os.path.join(os.getcwd(),'Data_processed\\Analysis - 1 _Data')
```

```python
# we are opening the text file to make use of the list of words  
file = open("warword.txt","r")
warwords=file.read() # we are reading the file 
warwords = warwords.lower() # we are lowering the words 
warwords = word_tokenize(str(warwords))# we are tokenizing the words 
            
```

```python
#We are defining the empty list and dictionary 
count = 0
worddict={}
xaxis=[]
yaxis=[]
```

```python

for name1 in os.listdir(filepath):# It will list all the directory 
        newd1= os.path.join(filepath,name1) # it will join two path 
        for file in glob.glob(os.path.join( newd1,'*.json')): # it will join two path 
                with open(file) as json_file:
                    json_data = json.load(json_file) # reading the json data
                    location = json_data["lead_paragraph"]
                    justwords = word_tokenize(str(location))# tokenizer 
                    for words in justwords: # 
                            words= words.lower()
                            if words in warwords:
                                count=count+1

        with open("Warwords.csv","a+") as outputfile1: # we are wititng into a csv file 
            outputfile1.write(" for the following year : " + str(name1) +" the words related to war occured "+ "," +str(count) +"\n")

        
        
        yaxis.append(count)# we are appending the value for yaxis 
        name1= str(name1)
        xaxis.append(name1)# we are appending the value for yaxis 
        count = 0
        
            
```

```python
# We are ploting the graph 
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
y_pos = np.arange(len(xaxis))
 
plt.bar(y_pos, yaxis, align='center', alpha=0.5)
plt.xticks(y_pos, xaxis)
plt.ylabel('Words Related to war')
plt.title('Graph for the words Related to war over the period of time')
 
plt.show()
```
