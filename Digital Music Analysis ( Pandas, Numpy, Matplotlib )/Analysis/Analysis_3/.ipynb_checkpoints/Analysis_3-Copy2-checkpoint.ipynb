```python
import pandas as pd
import os
from operator import itemgetter # Importing the itemgetter
from datetime import datetime
from dateutil.parser import parse
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import mpld3
```

```python
Root_path=os.getcwd()
data= "Data"
Data= os.path.join(Root_path,data) # we are joining the old with new value to form a new path 
```

```python
path_album_details="Spotify_Album_complete_details.csv"
Data_album_details= os.path.join(Data,path_album_details) # we are joining the old with new value to form a new path 

path_twitter_details="Twitter_details.csv"
Data_twitter_details= os.path.join(Data,path_twitter_details) # we are joining the old with new value to form a new path 
```

```python
Album_details = pd.DataFrame.from_csv(Data_album_details, encoding = "ISO-8859-1").reset_index()
```

```python
LI= Album_details.Album_release_date
```

```python

def try_parsing_date(text):
    for fmt in ('%Y-%m-%d','%d-%m-%Y','%m-%d-%Y', '%Y-%d-%m', '%Y', '%Y-%m', '%Y-%d'):
        try:
            datas = datetime.strptime(text, fmt).strftime('%Y')
            return datas
        except ValueError:
            pass
    raise ValueError('no valid date format found')
```

```python
DATES=[]
for i in range (0,len(LI)):
    date=LI[i]
    DATES.append(try_parsing_date(date))
```

```python
Album_details["YEAR"]=DATES
```

```python
Album_details["YEAR"]= Album_details["YEAR"].astype(int)
```

```python
Start_year = input('Enter a Start_year in YYYY format only: ')
End_year = input('Enter a End_year in YYYY format only : ')

if Start_year < End_year:
    yearlist = list (range (int(Start_year),int(End_year)))
    Filtered_with_Year= pd.DataFrame(Album_details[Album_details['YEAR'].isin(yearlist)]).reset_index()
    Filtered_with_Year_max = pd.DataFrame( Filtered_with_Year.groupby("YEAR",sort=True).Album_popularity.max().reset_index())
    Filtered_with_Year_copy =Filtered_with_Year[["Album_name","Artist1_name","YEAR","Album_popularity","Album_images"]].copy()
    Filtered_with_merged=pd.merge(Filtered_with_Year_max,Filtered_with_Year_copy, on=("YEAR","Album_popularity"))
else:
    print("The year entered is not in proper format or in range ")
    
    
```

```python
max_max = pd.DataFrame( Filtered_with_merged[ (Filtered_with_merged.Album_popularity== Filtered_with_merged.Album_popularity.max())].reset_index())
```

```python
# plotting the rate of increase or decrease in crimes
Album_popularity = Filtered_with_merged['Album_popularity']
TITLE = Filtered_with_merged['Album_&YEAR']

fig = plt.figure(figsize=(8,6))
fig.add_subplot(2,1,1)

x = np.arange(len(TITLE))
bar_width = 0.2

plt.bar(x, Album_popularity,bar_width, color='#1a1aff')
plt.legend()
plt.xlabel('Album Name')
plt.ylabel('Album popularity')
plt.title("Top 5 albums and their popularity for the given year range")
plt.xticks(x+bar_width,TITLE, rotation=30)
plt.tight_layout()

```

# does the popularity depends on the artist 

```python
twitter_details = pd.DataFrame.from_csv(Data_twitter_details, encoding = "ISO-8859-1").reset_index()
```

```python
name_list=[]
for names in Filtered_with_merged.Artist1_name:
    formated_name= str(names).replace("'","").replace("$","s").replace("!","").replace("&","")
    name_list.append(formated_name)
```

```python
# we are extracting the month from date 
twitter_details["Count"]=1
```

```python
Value_matching_= pd.DataFrame(twitter_details[twitter_details['searchname'].isin(name_list)])
    
```

```python
Artist_tweets= pd.DataFrame( Value_matching_.groupby("searchname",sort=True).retweet_count.sum().reset_index())
```

```python
Artist_tweets=Artist_tweets.rename(columns = {'searchname':'Artist1_name'})
```

```python
final_table= pd.merge(Artist_tweets,Filtered_with_merged, on="Artist1_name")
```

```python

final_table["Album_By_artist"]=  final_table[['Album_name', 'Artist1_name']].apply(lambda x: '  - (by) -  '.join(x), axis=1)
```

```python
final_table
```

```python
final_table_retweet_count= final_table["retweet_count"].sum()
final_table["Tweets_in_percent"] = round((final_table["retweet_count"]/final_table_retweet_count)*100)
popularity_sum= final_table["Album_popularity"].sum()
final_table["Album_popularity_in_percent"] = round((final_table["Album_popularity"]/popularity_sum)*100)
```

```python

# plotting the rate of increase or decrease in crimes
Tweets_in_percent = final_table["Tweets_in_percent"]
Album_popularity_in_percent = final_table["Album_popularity_in_percent"] 
TITLE = final_table['Album_By_artist']

fig = plt.figure(figsize=(8,6))
fig.add_subplot(2,1,1)

x = np.arange(len(TITLE))
bar_width = 0.2
opacity = 0.5 #not so dark

plt.bar(x, Tweets_in_percent,bar_width, color='#ffff00', label='Tweets(percent)')
plt.bar(x+bar_width, Album_popularity_in_percent,bar_width, color='#ff00ff', label='Album_popularity(percent)')
plt.legend()
plt.xlabel('Album and Artist Name')
plt.ylabel('popularity count (percentage)')
plt.title("Album popularity and Artist popularity")
plt.xticks(x+bar_width,TITLE, rotation=30)
plt.tight_layout()

```

```python
tweet_pasific_time= pd.DataFrame(Value_matching_[Value_matching_['time_zone'].isin(just_timezone)]).reset_index()
```

```python
tweet_timezone= pd.DataFrame( tweet_pasific_time.groupby("time_zone",sort=True).sum().reset_index())
```

```python
tweet_timezone_copy =tweet_timezone[["time_zone","statuses_count","friends_count","followers_count"]].copy()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
friends_count = list(tweet_timezone_copy['friends_count'])
statuses_count = list(tweet_timezone_copy['statuses_count'])

followers_count = list(tweet_timezone_copy['followers_count'])
time_zone = list(tweet_timezone_copy['time_zone'])

```

```python
def labeling(rects):      # Add top labels
    for rect in rects:
        height = rect.get_height()
        label_position = height * 1.02
        plt.text(rect.get_x() + rect.get_width()/2., label_position, '%d' % int(height), ha='center', va='bottom', fontsize=15)
```

```python
plt.subplots(figsize=(20,8))   
sns.set_style("darkgrid")

index = np.arange(5)
bar_width = 0.3
opacity = .3
b1 = plt.bar(index + bar_width/2, friends_count, bar_width,  color='#000000', label='Friends')
b2 = plt.bar(index + bar_width/2, followers_count, bar_width, alpha=opacity, color='#99ffff', label='Followers')
labeling(b1)
labeling(b2)

plt.title('Friend and follower (for users) in Top 5 timezone ', fontweight='bold', fontsize=35)     
plt.xlabel('Time_zone', fontsize=35)
plt.ylabel('Friend & Follower Count', fontsize=20)
plt.xticks(index + bar_width, time_zone, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=30)
plt.tight_layout()
plt.show()
```
