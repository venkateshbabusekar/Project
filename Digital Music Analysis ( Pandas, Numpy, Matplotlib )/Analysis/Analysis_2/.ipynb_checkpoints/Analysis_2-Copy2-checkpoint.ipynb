```python
Sentiment_datas = pd.DataFrame.from_csv(Sentiment_final_details, encoding = "ISO-8859-1").reset_index()
```

```python
Sentiment_data= pd.merge(Sentiment_datas,product_details_SAMPLE_tocompares, on="ASIN")
```

```python
# we are creating a lambda function 
Sentiment_data["Really_Helpfull"] = Sentiment_data["Helpful"].apply(lambda x:x.split("\\",1)[-2]).astype(int)
```

```python
Sentiment_data["Not_Helpfull"] = Sentiment_data["Helpful"].apply(lambda x:x.split("\\",1)[-1]).astype(int)
```

```python
#Sentiment_data=Sentiment_data.rename(columns = {"0":"Sentimental_score"})
```

```python
# We are creating a data frame where we finter the data for n
Helpful=pd.DataFrame( Sentiment_data.groupby(["TITLE","sentimental_score"],sort=False).Really_Helpfull.sum().reset_index())
```

```python
# We are creating a data frame where we finter the data for newyork 
NotHelpful =pd.DataFrame( Sentiment_data.groupby(["TITLE","sentimental_score"],sort=False).Not_Helpfull.sum().reset_index())
```

```python
Filtered_with_merged=pd.merge(NotHelpful,Helpful, on=("TITLE","sentimental_score"))
```

```python

# We are creating a data frame where we finter the data for newyork 
negative_sentiment_h_nh =pd.DataFrame( Filtered_with_merged[Filtered_with_merged.sentimental_score == 0.0].reset_index())
# We are creating a data frame where we finter the data for newyork 
positive_sentiment_h_nh =pd.DataFrame( Filtered_with_merged[Filtered_with_merged.sentimental_score == 1.0].reset_index())
```

```python
# plotting the rate of increase or decrease in crimes
Not_Helpfull = negative_sentiment_h_nh['Not_Helpfull']
Really_Helpfull = negative_sentiment_h_nh['Really_Helpfull']
TITLE = negative_sentiment_h_nh['TITLE']

fig = plt.figure(figsize=(8,6))
fig.add_subplot(2,1,1)

x = np.arange(len(TITLE))
bar_width = 0.2
opacity = 0.5 #not so dark

plt.bar(x, Not_Helpfull,bar_width, color='#000000', label='Not_Helpfull(reviews)')
plt.bar(x+bar_width, Really_Helpfull,bar_width, color='#00802b', label='Helpfull(reviews)')
plt.legend()
plt.xlabel('Product_Name(Digital music)')
plt.ylabel('Helpfull / Not_Helpfull (count)')
plt.title("Negative Review - Helpfulness for reviews for each product")
plt.xticks(x+bar_width,TITLE, rotation=30)
plt.tight_layout()



fig.add_subplot(2,1,2)
Not_Helpfull1 = positive_sentiment_h_nh['Not_Helpfull']
Really_Helpfull1 = positive_sentiment_h_nh['Really_Helpfull']

plt.bar(x, Not_Helpfull1,bar_width, color='#000000', label='Not_Helpfull(reviews)')
plt.bar(x+bar_width, Really_Helpfull1,bar_width, color='#9900cc', label='Helpfull(reviews)')
plt.legend()
plt.xlabel('Product_Name(Digital music)')
plt.ylabel('Helpfull / Not_Helpfull (count)')
plt.title("Positive Review - Helpfulness for reviews for each product")
plt.xticks(x+bar_width,TITLE, rotation=30)
plt.tight_layout()

plt.show()
```
