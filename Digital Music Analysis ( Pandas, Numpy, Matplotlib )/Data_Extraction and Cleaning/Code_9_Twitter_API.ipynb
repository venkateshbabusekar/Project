```python
# LOADING THE FIILE 
import pandas as pd
import datetime
import os
Full_details = pd.DataFrame.from_csv("Top_artist_name.csv", encoding = "ISO-8859-1").reset_index()
```

```python
Just_Name= Full_details[" Artist_name"].unique()
```

```python
# HITING THE API TO GET THE DATA 
for i in range(0,len(Just_Name)):
    query =Just_Name[i]
    query= str(query).replace("'","").replace("$","s").replace("!","").replace("&","")
    name = os.path.join(path_toread_Data,query) # we are joining the old with new value to form a new path 
    time=datetime.datetime.now()
    time=time.strftime("%y-%m-%d-%H")
    filename = name+time # we are joining the old with new value to form a new path 
    
    a =requests.get('https://api.twitter.com/1.1/search/tweets.json?q='+query+'&count=100&lang=en',auth=uauth).json()
    with open(filename+'.json', 'w') as outfile:
        json.dump(a, outfile)

```

```python
from os import listdir
from PIL import Image as PImage
import matplotlib.pyplot as plt; plt.rcdefaults()

def loadImages(path):
    # return array of images

    imagesList = listdir(path)
    loadedImages = []
    for image in imagesList:
        img = PImage.open(path + image)
        loadedImages.append(img)

    return loadedImages

path = "images/"

# your images in an array
imgs = loadImages(path)

for img in imgs:
    # you can show every image
    img.show()
    
%matplotlib inline
plt.imshow(img)
```
